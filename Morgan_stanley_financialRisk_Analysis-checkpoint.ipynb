### TASK 1: Data Cleaning & Formatting

```python
financial_cols = ["TransactionAmount", "AccountBalance"]

for col in financial_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(r"[₹$,]", "", regex=True)
        .astype(float)
    )
```

#### Step 1.2: Convert Transaction Date to Proper Format

```python
df["TransactionDate"] = pd.to_datetime(
    df["TransactionDate"],
    format="%d-%m-%Y",
    errors="coerce"
)
```

```python
df["TransactionDate"].isnull().sum()
```

#### Step 1.3: Standardize Categorical Columns

```python
df["AccountType"] = (
    df["AccountType"]
    .str.lower()
    .str.strip()
)
```

```python
df["TransactionType"] = (
    df["TransactionType"]
    .str.lower()
    .str.strip()
)
```

```python
df["TransactionType"] = df["TransactionType"].replace({
    "deposit": "credit",
    "withdrawal": "debit",
    "payment": "debit",
    "transfer": "debit"
})
```

#### Step 1.4: Handle Missing Values

```python
df.isnull().sum()
```

```python
yearly_summary["Netflow"] = (yearly_summary.get("credit",0) - yearly_summary.get("debit",0))
```

### Task 3: Customer Profile Building

#### Step 3.1: Create Account-Level Summary

```python
account_summary = (
    df
    .groupby("AccountID")
    .agg(
        total_transactions=("TransactionID", "count"),
        total_amount=("TransactionAmount", "sum"),
        avg_balance=("AccountBalance", "mean"),
        credit_count=("TransactionType", lambda x: (x == "credit").sum()),
        debit_count=("TransactionType", lambda x: (x == "debit").sum())
    )
    .reset_index()
)
```

```python
account_summary.head()
```

#### STEP 3.2: Normalize Activity (Transactions per Month)

```python
total_months = df["Month"].nunique()

account_summary["txn_per_month"] = (
    account_summary["total_transactions"] / total_months
)
```

```python
account_summary["txn_per_month"].describe()
```

#### Step 3.3: Define Activity Level Rubric 

```python
high_cutoff = account_summary["txn_per_month"].quantile(0.75)
low_cutoff = account_summary["txn_per_month"].quantile(0.25)

def activity_level(txn_rate):
    if txn_rate >= high_cutoff:
        return "High"
    elif txn_rate < low_cutoff:
        return "Low"
    else:
        return "Medium"

account_summary["ActivityLevel"] = (
    account_summary["txn_per_month"]
    .apply(activity_level)
)
```

```python
account_summary["ActivityLevel"].value_counts()
```

#### STEP 3.4: Segment Accounts by Balance

```python
account_summary["BalanceSegment"] = pd.qcut(
    account_summary["avg_balance"],
    q=3,
    labels=["Low Balance", "Medium Balance", "High Balance"]
)
```

#### STEP 3.5: Segment Accounts by Transaction Volume

```python
account_summary["TransactionVolumeSegment"] = pd.qcut(
    account_summary["total_amount"],
    q=3,
    labels=["Low Volume", "Medium Volume", "High Volume"]
)
```

#### Profile 1: High Net Inflow Accounts

```python
account_summary = account_summary.merge(
    account_flow[["Net_Inflow"]],
    left_on="AccountID",
    right_index=True,
    how="left"
)
```

```python
high_net_inflow_accounts = account_summary[
    (account_summary["Net_Inflow"] > 0) &
    (account_summary["TransactionVolumeSegment"] == "High Volume")
]
```

### TASK 4: Financial Risk Identification

#### STEP 4.1: Identify Accounts with Frequent Large Withdrawals

```python
large_withdrawal_threshold = df[
    df["TransactionType"] == "debit"
]["TransactionAmount"].quantile(0.90)

large_withdrawal_threshold
```

#### Step 2: Count large withdrawals per account

```python
large_withdrawals = df[
    (df["TransactionType"] == "debit") &
    (df["TransactionAmount"] >= large_withdrawal_threshold)
]

large_withdrawal_counts = (
    large_withdrawals
    .groupby("AccountID")
    .size()
    .reset_index(name="large_withdrawal_count")
)
```

#### Step 3: Flag risky accounts

```python
risky_withdrawal_accounts = large_withdrawal_counts[
    large_withdrawal_counts["large_withdrawal_count"] >= 3
]

risky_withdrawal_accounts.head()
```

Rubric :
Accounts with 3 or more large withdrawals were flagged as high withdrawal risk.

#### STEP 4.2: Measure Balance Volatility (Stability Risk)

##### Step 1: Calculate balance volatility (standard deviation)

```python
balance_volatility = (
    df
    .groupby("AccountID")["AccountBalance"]
    .std()
    .reset_index(name="balance_std")
)
```

##### Step 2: Flag highly volatile accounts

```python
volatility_threshold = balance_volatility["balance_std"].quantile(0.75)

high_volatility_accounts = balance_volatility[
    balance_volatility["balance_std"] >= volatility_threshold
]

high_volatility_accounts.head()
```

Rubric:

Accounts in the top 25% of balance volatility were flagged as unstable.

#### STEP 4.3: Detect Anomalous Transactions (IQR Method)

##### Step 1: Compute IQR bounds

```python
Q1 = df["TransactionAmount"].quantile(0.25)
Q3 = df["TransactionAmount"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

##### Step 2: Flag anomalous transactions

```python
anomalous_transactions = df[
    (df["TransactionAmount"] < lower_bound) |
    (df["TransactionAmount"] > upper_bound)
]
```

##### Step 3: Identify accounts with frequent anomalies

```python
anomaly_counts = (
    anomalous_transactions
    .groupby("AccountID")
    .size()
    .reset_index(name="anomaly_count")
)

high_anomaly_accounts = anomaly_counts[
    anomaly_counts["anomaly_count"] >= 3
]

high_anomaly_accounts.head()
```

Rubric:
Accounts with three or more anomalous transactions were flagged as irregular.

#### STEP 4.4: Consolidate Financial Risk Flags

```python
risk_profile = (
    account_summary
    .merge(risky_withdrawal_accounts, on="AccountID", how="left")
    .merge(high_volatility_accounts, on="AccountID", how="left")
    .merge(high_anomaly_accounts, on="AccountID", how="left")
)
```

```python
risk_profile[["large_withdrawal_count", "balance_std", "anomaly_count"]] = (
    risk_profile[["large_withdrawal_count", "balance_std", "anomaly_count"]]
    .fillna(0)
)  ## Fill missing flags
```

```python
risk_profile["RiskFlag"] = (
    (risk_profile["large_withdrawal_count"] >= 3) |
    (risk_profile["balance_std"] >= volatility_threshold) |
    (risk_profile["anomaly_count"] >= 3)
)  ## Create overall risk flag
```

### TASK 6: Hypothesis Testing

#### TASK 6.1: Define Hypotheses

Let:

Group 1 = High Volume Accounts

Group 2 = Low Volume Accounts

🔹 Null Hypothesis (H₀):

There is no difference in average balance between high-volume and low-volume accounts.

🔹 Alternative Hypothesis (H₁):

High-volume accounts have higher average balances than low-volume accounts.

#### TASK 6.2: Extract the Two Groups

```python
high_volume_balances = account_summary[
    account_summary["TransactionVolumeSegment"] == "High Volume"
]["avg_balance"]

low_volume_balances = account_summary[
    account_summary["TransactionVolumeSegment"] == "Low Volume"
]["avg_balance"]
```

```python
len(high_volume_balances), len(low_volume_balances)
```

#### TASK 6.3: Perform Independent T-Test

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(
    high_volume_balances,
    low_volume_balances,
    equal_var=False 
)

t_stat, p_value
```

#### TASK 6.4: Interpret the Result

```python
alpha = 0.05
if p_value < alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

```python
plt.figure(figsize=(6,4))

plt.boxplot(
    [high_volume_balances, low_volume_balances],
    labels=["High Volume", "Low Volume"]
)

plt.title("Average Balance Comparison by Transaction Volume")
plt.ylabel("Average Balance")
plt.tight_layout()
plt.show()
```
